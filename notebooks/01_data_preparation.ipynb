# 01 - Przygotowanie danych

Ten notebook przygotowuje korpus danych do budowy datasetu RAFT:
- **Golden Documents (D*):** artykuły fact-checkingowe, komunikaty WHO/PZH
- **Dystraktory (Dᵢ):** teksty spiskowe, fake news

```python
import sys
sys.path.insert(0, "../src")

from data_collection import (
    SAMPLE_GOLDEN_DOCS,
    SAMPLE_DISTRACTORS,
    save_documents,
    create_manual_golden_doc,
    create_manual_distractor,
    scrape_konkret24_article,
    batch_scrape,
)
```

## 1. Załaduj przykładowe dane (quick-start)

```python
# Użyj wbudowanych przykładów jako punkt wyjścia
golden_docs = list(SAMPLE_GOLDEN_DOCS)
distractors = list(SAMPLE_DISTRACTORS)

print(f"Golden docs: {len(golden_docs)}")
print(f"Dystraktory: {len(distractors)}")
```

## 2. Rozszerzenie danych (opcjonalnie)

### 2a. Dodawanie ręczne

```python
# Dodaj własne golden docs
golden_docs.append(create_manual_golden_doc(
    title="Maski ochronne - skuteczność",
    content=(
        "Według metaanalizy opublikowanej w British Medical Journal (BMJ), "
        "maski chirurgiczne i N95/FFP2 zmniejszają ryzyko zakażenia wirusami "
        "dróg oddechowych o 14-27%. Mechanizm działania polega na fizycznej "
        "filtracji kropelek i aerozoli. Maski są szczególnie skuteczne w "
        "zamkniętych pomieszczeniach z ograniczoną wentylacją. WHO rekomenduje "
        "noszenie masek w przestrzeniach publicznych podczas pandemii."
    ),
    source="BMJ/WHO",
))

# Dodaj własne dystraktory
distractors.append(create_manual_distractor(
    content=(
        "Maski nie działają — wirus jest 1000x mniejszy od porów w masce! "
        "To tylko narzędzie kontroli społecznej, żeby nas przyzwyczaić do "
        "posłuszeństwa. Kraje bez obowiązku masek miały takie same wyniki "
        "jak te z obowiązkiem. Oddychanie CO2 przez maskę osłabia nasz "
        "układ odpornościowy i powoduje hipoksję. Lekarze wiedzą, ale boją "
        "się mówić, bo stracą pracę!"
    ),
    topic="maski",
))

print(f"Golden docs: {len(golden_docs)}")
print(f"Dystraktory: {len(distractors)}")
```

### 2b. Scraping fact-checków (opcjonalnie)

```python
# Opcjonalnie: scraping artykułów z Konkret24
# Wymaga manualnego podania URLi (lista poniżej to przykłady)

KONKRET24_URLS = [
    # Dodaj prawdziwe URLe artykułów fact-checkingowych o COVID/szczepionkach
    # Przykład: "https://konkret24.tvn24.pl/zdrowie,110/artykul-o-szczepionkach,12345.html"
]

if KONKRET24_URLS:
    scraped = batch_scrape(KONKRET24_URLS, scrape_konkret24_article, delay=3.0)
    golden_docs.extend(scraped)
    print(f"Scraped: {len(scraped)} artykułów")
else:
    print("Brak URLi do scrapingu. Uzupełnij listę KONKRET24_URLS.")
```

## 3. Zapis danych

```python
save_documents(golden_docs, "../data/golden_docs.jsonl")
save_documents(distractors, "../data/distractors.jsonl")

print("\nDane zapisane. Przejdź do notebook 02_dataset_generation.")
```
